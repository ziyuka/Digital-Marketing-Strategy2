Outliers do not necessarily need to be removed, especially if they represent valid data. #### **Message: Length** - Saharla

We want to add 3 new columns to analyze the length of the text. The new columns are going to be low, medium and high. In the literature research we found out that characters that are longer than 80 characters were more likely to have a higher level of user engagement (Gkikas et al., 2022). Therefore, Gkikas et al. (2022) adopted an 80-character threshold as a key indicator of user engagement and organized the texts into 80-character intervals: 0-80, 81-160, 161-240, 241-320, 321-400, 401-480, and 481-500. So anything less than 160 will be low. Anything between 160 and 320 will be medium and above 320 will be high.

```python
# Define a function to categorize posts based on length
def categorize_length(length):
    if length <= 160:
        return 'Postlength Low'
    elif length <= 320:
        return 'Postlength Medium'
    else:
        return 'Postlength High'

# Calculate character length of each post
merged_themepark['char_length'] = merged_themepark['Message'].apply(len)

# Categorize posts based on character length
merged_themepark['post_length_category'] = merged_themepark['char_length'].apply(categorize_length)

# Check the distribution of categories
print(merged_themepark['post_length_category'].value_counts())

# One-hot encode the 'post_length_category' into separate columns
merged_themepark['Postlength Low'] = (merged_themepark['post_length_category'] == 'Postlength Low').astype(int)
merged_themepark['Postlength Medium'] = (merged_themepark['post_length_category'] == 'Postlength Medium').astype(int)
merged_themepark['Postlength High'] = (merged_themepark['post_length_category'] == 'Postlength High').astype(int)
```

```python
import re

# Function to check for emojis using a regular expression
def contains_emoji(s):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return re.search(emoji_pattern, s) is not None

# Function to check for informal indicators
def is_informal(text):
    # Check for emojis
    if contains_emoji(text):
        return True

    # Check for abbreviations (e.g., "lol", "brb", "btw", "omg")
    if re.search(r'\b(?:lol|brb|btw|omg)\b', text, re.IGNORECASE):
        return True

    # Check for hashtags
    if '#' in text:
        return True

    # Check for colloquial language
    if re.search(r'\b(?:gonna|wanna|ain\'t)\b', text, re.IGNORECASE):
        return True

    # Check for exclamation marks
    if '!' in text:
        return True

    # If none of the informal indicators are present
    return False
```

```python
# Now you can apply this function to your DataFrame as shown earlier:
merged_themepark['Classification'] = merged_themepark['Message'].apply(is_informal).map({True: 'Informal', False: 'Not Informal'})
merged_themepark.shape
(15443, 18)
# Assuming 'Classification' column already exists from previous steps
# Create 'Is_Informal' column
merged_themepark['Is_Informal'] = merged_themepark['Classification'].apply(lambda x: 1 if x == 'Informal' else 0)

# Create 'Is_Not_Informal' column
merged_themepark['Is_Not_Informal'] = merged_themepark['Classification'].apply(lambda x: 1 if x == 'Not Informal' else 0)

# Display the DataFrame to verify the new columns
# merged_themepark

pd.set_option('display.max_rows', None)  # This will display all rows in the DataFrame

merged_themepark.head()
```

```python
# Assuming 'Classification' column already exists from previous steps# Create 'Is_Informal' columnmerged_themepark['Is_Informal'] = merged_themepark['Classification'].apply(lambda x: 1 if x == 'Informal' else 0)# Create 'Is_Not_Informal' columnmerged_themepark['Is_Not_Informal'] = merged_themepark['Classification'].apply(lambda x: 1 if x == 'Not Informal' else 0)# Display the DataFrame to verify the new columns# merged_themeparkpd.set_option('display.max_rows', None)  # This will display all rows in the DataFramemerged_themepark
import matplotlib.pyplot as plt

from scipy import stats
 
# Function to plot histogram and Q-Q plot and perform Shapiro-Wilk test

def analyze_normality(data, title):

    # Histogram

    plt.hist(data, bins='auto', alpha=0.7)

    plt.title(f'Histogram of {title}')

    plt.show()
 
    # Q-Q plot

    stats.probplot(data, dist="norm", plot=plt)

    plt.title(f'Q-Q Plot of {title}')

    plt.show()
 
    # Shapiro-Wilk test

    shapiro_test = stats.shapiro(data.dropna())

    print(f"Shapiro-Wilk Test for {title}: Statistic={shapiro_test[0]}, P-value={shapiro_test[1]}")

    if shapiro_test[1] > 0.05:

        print(f"The data in {title} is normally distributed (Fail to reject H0).\n")

    else:

        print(f"The data in {title} is not normally distributed (Reject H0).\n")
 
# Analyze 'Number of Comments' for 'Informal' and 'Not Informal'

analyze_normality(merged_themepark[merged_themepark['Is_Informal'] == 1]['Number of Comments'], 'Number of Comments for Informal Posts')

analyze_normality(merged_themepark[merged_themepark['Is_Not_Informal'] == 1]['Number of Comments'], 'Number of Comments for Not Informal Posts')
 
# Analyze 'Number of Likes' for 'Informal' and 'Not Informal'

analyze_normality(merged_themepark[merged_themepark['Is_Informal'] == 1]['Number of Likes'], 'Number of Likes for Informal Posts')

analyze_normality(merged_themepark[merged_themepark['Is_Not_Informal'] == 1]['Number of Likes'], 'Number of Likes for Not Informal Posts')
 
# Analyze 'Number of Shares' for 'Informal' and 'Not Informal'

analyze_normality(merged_themepark[merged_themepark['Is_Informal'] == 1]['Number of Shares'], 'Number of Shares for Informal Posts')

analyze_normality(merged_themepark[merged_themepark['Is_Not_Informal'] == 1]['Number of Shares'], 'Number of Shares for Not Informal Posts')
```

```python
from scipy.stats import ttest_ind

# Function to perform t-test and print the results
def perform_ttest(group1, group2, variable_name):
    # Perform the t-test
    stat, p = ttest_ind(group1.dropna(), group2.dropna())
   
    # Print the test results
    print(f"T-Test for {variable_name}:")
    print(f"T-statistic: {stat}, P-value: {p}\n")
    if p < 0.05:
        print(f"There is a significant difference in the mean number of {variable_name.lower()} between Informal and Not Informal posts.\n")
    else:
        print(f"No significant difference in the mean number of {variable_name.lower()} between Informal and Not Informal posts.\n")

# T-Test for 'Number of Comments'
perform_ttest(
    merged_themepark[merged_themepark['Is_Informal'] == 1]['Number of Comments'],
    merged_themepark[merged_themepark['Is_Not_Informal'] == 1]['Number of Comments'],
    'Comments'
)

# T-Test for 'Number of Likes'
perform_ttest(
    merged_themepark[merged_themepark['Is_Informal'] == 1]['Number of Likes'],
    merged_themepark[merged_themepark['Is_Not_Informal'] == 1]['Number of Likes'],
    'Likes'
)

# T-Test for 'Number of Shares'
perform_ttest(
    merged_themepark[merged_themepark['Is_Informal'] == 1]['Number of Shares'],
    merged_themepark[merged_themepark['Is_Not_Informal'] == 1]['Number of Shares'],
    'Shares'
)
```
